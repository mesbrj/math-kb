```python
from IPython.display import display, Latex
from sympy import symbols, lambdify, plot
from sympy.plotting import plot3d
import matplotlib.pyplot as plt

```

```python
# Mathematical expression
expression_result = 2 * ((3+2)**2 / 5) - 4
latex_expression = f"$$ 2\\times \\frac{{(3 + 2)^2}}{{5}} - 4 = {int(expression_result)}$$"
display(Latex(latex_expression))
```

```python
# Linear function
latex_function = f"$$f(x) = 2x + 1$$"
display(Latex(latex_function))

# Plot the Linear function using SymPy
x_var = symbols('x')
f_func = 2 * x_var + 1
plot(f_func, title='', xlabel='', ylabel='')
```

```python
# Evaluate the function at a specific point (from standard Python function)
latex_function = f"$f(x) = 2x + 1$"
display(Latex(latex_function))

def f(x):
    return 2 * x + 1

x, y = 5, f(5)
print(f"x = {x}, f(x) = {y}")

# Plotting the point on the function graph using Matplotlib
x_vals = [i/10 for i in range(0, 121)]
y_vals = [f(x) for x in x_vals]
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = 2x + 1') # Function line
plt.plot(x, y, 'ro', markersize=10, label=f'x={x}, f({x})={y}') # Point
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.title('Linear Function with Point (using Standard Python Function)')
plt.show()
```

```python
# Evaluate the function at a specific point (from SymPy substitution)
latex_function = f"$f(x) = 2x + 1$"
display(Latex(latex_function))

x_var = symbols('x')
func = 2*x_var + 1
x = 7
y = func.subs(x_var, x)
print(f"x = {x}, f(x) = {y}")

# Plotting the point on the function graph using Matplotlib
f_numeric = lambdify(x_var, func) # Convert symbolic function to numerical for plotting
x_vals = [i/10 for i in range(0, 121)]
y_vals = [f_numeric(x) for x in x_vals]
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = 2x + 1') # Function line
plt.plot(x, y, 'ro', markersize=10, label=f'Point ({x}, {y})') # Point
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.title('Linear Function with Point (using SymPy substitution)')
plt.show()
```

```python
# multivariable linear function
latex_function = f"$f(x, y) = 2x + 3y$"
display(Latex(latex_function))

# 3D Plot the multivariable function using SymPy
x, y = symbols('x y') 
f = 2*x + 3*y 
plot3d(f)
```

```python
# Quadratic function
latex_function = f"$$f(x) = x^2 + 4$$"
display(Latex(latex_function))

x = symbols('x')
f = x**2 + 4
plot(f, title='', xlabel='', ylabel='')
```

```python
# Evaluate the function at a specific point (from standard Python function)
latex_function = f"$$f(x) = x^2 + 4$$"
display(Latex(latex_function))

def f_quad(x):
    return x**2 + 4

x, y = 3, f_quad(3)
print(f"x = {x}, f(x) = {y}")

# Plotting the point on the function graph using Matplotlib
x_vals = [i/10 for i in range(-50, 51)]
y_vals = [f_quad(x) for x in x_vals]
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = x^2 + 4') # Function line
plt.plot(x, y, 'ro', markersize=10, label=f'x={x}, f({x})={y}') # Point
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.title('Quadratic Function with Point (using Standard Python Function)')
plt.show()
```
