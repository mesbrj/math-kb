```python
from IPython.display import display, Latex
from math import log
```

```python
display(Latex(f"$$a^x = b \\iff \\log_a b = x$$"))

print("The logarithm of 8 with base 2")
x = log(8, 2)
display(Latex(f"$$\\log_2 8 = {x}$$"))

print("The logarithm of 8 with base 10")
x = log(8, 10)
display(Latex(f"$$\\log_{{10}} 8 = {x}$$"))

print("Natural Logarithm (python std lib) with default base (Euler's number)")
x = log(8)
display(Latex(f"$$\\log 8 = {x}$$"))
x = log(10)
display(Latex(f"$$\\log 10 = {x}$$"))
```
