```python
from IPython.display import display, Latex
from sympy import symbols, lambdify, plot, Sum
from sympy.plotting import plot3d
import matplotlib.pyplot as plt
```

```python
# Linear function
latex_function = f"$$f(x) = 2x + 1$$"
display(Latex(latex_function))

# Plot the Linear function using SymPy
x_var = symbols('x')
f_func = 2 * x_var + 1
plot(f_func, title='', xlabel='', ylabel='')
```

```python
# Evaluate the function at a specific point (from standard Python function)
latex_function = f"$f(x) = 2x + 1$"
display(Latex(latex_function))

def f(x):
    return 2 * x + 1

x, y = 5, f(5)
print(f"({x}, {y})")

# Plotting the point on the function graph using Matplotlib
x_vals = [i/10 for i in range(0, 121)]
y_vals = [f(x) for x in x_vals]
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = 2x + 1') # Function line
plt.plot(x, y, 'ro', markersize=10, label=f'({x}, {y})') # Point
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.title('Linear Function with Point (using Standard Python Function)')
plt.show()
```

```python
# Evaluate the function at a specific point (from SymPy substitution)
latex_function = f"$f(x) = 2x + 1$"
display(Latex(latex_function))

x_var = symbols('x')
func = 2*x_var + 1
x = 7
y = func.subs(x_var, x)
print(f"({x}, {y})")

# Plotting the point on the function graph using Matplotlib
f_numeric = lambdify(x_var, func) # Convert symbolic function to numerical for plotting
x_vals = [i/10 for i in range(0, 121)]
y_vals = [f_numeric(x) for x in x_vals]
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = 2x + 1') # Function line
plt.plot(x, y, 'ro', markersize=10, label=f'Point ({x}, {y})') # Point
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.title('Linear Function with Point (using SymPy substitution)')
plt.show()
```

```python
# Evaluate the function at multiple points
latex_function = f"$$f(x) = - \\frac{{2}}{{3}}x + 5$$"
display(Latex(latex_function))

def f(x):
    return - (2/3) * x + 5

x_vars = [-2, 0, 3, 6, 9, 15, 18]
dependent_vars = []
for x in x_vars:
    dependent_vars.append(f(x))
    print(f"({x}, {dependent_vars[-1]})")

# Classic cartesian plane
fig, ax = plt.subplots(figsize=(8, 6))
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_xlabel('x', loc='right')
ax.set_ylabel('f(x)', loc='top')
ax.grid(True)
# Plot the function and points
x_vals = [i/10 for i in range(-200, 200)]
y_vals = [f(x) for x in x_vals]
ax.plot(x_vals, y_vals, 'b-', linewidth=2) # Function line
for x, y in zip(x_vars, dependent_vars):
    ax.plot(x, y, 'ro', markersize=10) # Points
plt.title('Linear Function with Multiple Points')
plt.show()
```

```python
# multivariable linear function
latex_function = f"$f(x, y) = 2x + 3y$"
display(Latex(latex_function))

# 3D Plot the multivariable function using SymPy
x, y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```
