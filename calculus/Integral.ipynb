```python
from IPython.display import display, Latex
import matplotlib.pyplot as plt
from sympy import latex, symbols
```

```python
print("Linear function:")
display(Latex("$f(x) = 2x$"))

def f(x):
    return 2 * x

print("Finding the area bounded between the line and the x-axis in the x range 0.0 to 1.0")
# Calculate area using triangle formula
base = 1
height = f(1)
area = 0.5 * base * height

display(Latex("$A = \\frac{1}{2}bh $"))
display(Latex(f"$A = \\frac{{1}}{{2}} \\times {base} \\times {height} = {area}$"))

fig, ax = plt.subplots(figsize=(8, 6))
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True, alpha=0.3)
ax.set_xlim(-1, 1.2)
ax.set_ylim(-1, 2.5)
ax.set_xlabel('x', loc='right', fontsize=12)
ax.set_ylabel('f(x)', loc='top', fontsize=12)
x_vals = [i / 100 for i in range(101)]
y_vals = [f(x) for x in x_vals]
ax.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = 2x')
ax.fill_between(x_vals, 0, y_vals, alpha=0.3, color='blue', label=f'Area = {area}')
ax.legend(fontsize=10)
ax.set_title('Area Under Linear Function f(x) = 2x from x=0 to x=1', fontsize=14)
plt.show()

```

```python
# Nonlinear function with Riemann sums

print("Nonlinear function:")
display(Latex("$f(x) = x^2 + 1$"))

def g(x):
    return x**2 + 1

print("Finding the area bounded between the curve and the x-axis in the x range 0.0 to 1.0")

# Calculate area using Riemann sums (numerical approximation)
n_rectangles = 5
width = 1.0 / n_rectangles
riemann_sum = sum(g(i * width) * width for i in range(n_rectangles))

display(Latex(f"Riemann sum approximation (n={n_rectangles}): $A \\approx {riemann_sum:.6f}$"))

# Calculate exact area using sympy integration
from sympy import integrate
x = symbols('x')
f_sym = x**2 + 1
exact_area = integrate(f_sym, (x, 0, 1))
display(Latex(f"Exact area: $A = \\int_0^1 (x^2 + 1) dx = {latex(exact_area)} = {float(exact_area):.6f}$"))


fig, ax = plt.subplots(figsize=(8, 6))
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True, alpha=0.3)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1, 2.5)
ax.set_xlabel('x', loc='right', fontsize=12)
ax.set_ylabel('f(x)', loc='top', fontsize=12)

# Plot the filled area under the curve
x_vals_area = [i / 100 for i in range(101)]
y_vals_area = [g(x) for x in x_vals_area]
ax.fill_between(x_vals_area, 0, y_vals_area, alpha=0.2, color='red', label=f'Exact Area = {float(exact_area):.3f}')

# Plot the rectangles for Riemann sum visualization
for i in range(n_rectangles):
    x_left = i * width
    height_rect = g(x_left)  # left Riemann sum (using left endpoint)
    ax.bar(x_left + width/2, height_rect, width=width, alpha=0.4, 
           color='orange', edgecolor='black', linewidth=0.5, align='center')

# Plot the function curve
x_vals = [i / 100 for i in range(-150, 151)]
y_vals = [g(x) for x in x_vals]
ax.plot(x_vals, y_vals, 'r-', linewidth=2, label='f(x) = x + 1')

ax.legend(fontsize=10)
ax.set_title(f'Riemann Sum with {n_rectangles} rectangles (Area  {riemann_sum:.3f})', fontsize=14)
plt.show()
```
