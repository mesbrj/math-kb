```python
from IPython.display import display, Latex
import matplotlib.pyplot as plt
from sympy import latex, symbols, diff, limit
from sympy.plotting import plot3d
```

```python
# Derivatives

def f(x):
    return x**2

display(Latex("$ y = f(x) = x^2$"))

def slope(f, x):
    m = (f(x[1]) - f(x[0])) / (x[1] - x[0])
    return m

x_vars = [2 , 2.1]
dependent_vars = [f(x) for x in x_vars]

for x, y in zip(x_vars, dependent_vars):
    print(f"({x}, {y})")

m = slope(f, x_vars)
display(Latex(
    f"The slope (m) between $x = {x_vars[0]:.2f}$"
    f" and $x = {x_vars[1]:.2f}$ is $m = {m:.2f}$")
)

fig, ax = plt.subplots(figsize=(8, 6))
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_xlabel('x', loc='right')
ax.set_ylabel('f(x)', loc='top')
ax.grid(True)
# function curve values
x_vals = [i / 10.0 for i in range(-25, 26)]
y_vals = [f(i) for i in x_vals]
# secant line values
# compute (b) y-intercept when x=0 (Intercept Form): b = y - mx
y_intercept = dependent_vars[0] - m * x_vars[0]
# secant line equation (Point-Slope Form): y = mx + b
x_secant = [x for x in x_vals if m * x + y_intercept >= 0]
y_secant = [m * x + y_intercept for x in x_secant]
# Plotting
ax.plot(x_vals, y_vals, label='f(x) = x^2')
ax.plot(x_vars, dependent_vars, 'ro')
ax.plot(x_secant, y_secant, 'r--', label='Secant Line')
ax.legend()
plt.title('Function and Secant Line (slope visualization)')
plt.show()

# compute the derivative using SymPy:

print("using SymPy to find the derivative applying the power rule:")
x = symbols('x')
f_sym = x**2
f_prime = diff(f_sym, x)
display(Latex(f"$ \\frac{{d}}{{dx}} f(x) = \\frac{{d}}{{dx}} x^2 = {latex(f_prime)} $"))


display(Latex(f"The value at $x = 4$ is $f(4) = {f(4)}$"))
display(Latex(f"The derivative value at $x = 4$ is $f'(4) = {f_prime.subs(x, 4)}$"))

fig, ax = plt.subplots(figsize=(8, 6))
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_xlabel('x', loc='right')
ax.set_ylabel('f(x)', loc='top')
ax.grid(True)
x_vals = [i / 10.0 for i in range(-45, 46)]
y_vals = [f(i) for i in x_vals]
y_vals_derivative = [f_prime.subs(x, i) for i in x_vals]
ax.plot(x_vals, y_vals, label='f(x) = x^2')
ax.plot(x_vals, y_vals_derivative, label="f'(x) = 2x")
ax.plot([4], [f(4)], 'ro')
ax.plot([4], [f_prime.subs(x, 4)], 'bo')
ax.legend()
plt.title('Function and its Derivative')
plt.show()

```

```python
# Limits to calculate derivatives

display(Latex(f"$ y = f(x) = x^3 $"))
def new_func(x):
    return x**3
x_vars = [2, 2.1, 2.01, 2.00001]
y_vars = [new_func(x) for x in x_vars]

for x, y in zip(x_vars, y_vars):
    print(f"({x}, {y})")

display(Latex(f"$\\lim_{{h \\to 0}} \\frac{{f(x + h) - f(x)}}{{h}}$"))

print(
    "When x = 2, f(2) = 8\n"
    "When h = 0.1:    slope  (9.261 - 8) / 0.1 = 12.61\n"
    "When h = 0.01:   slope  (8.120 - 8) / 0.01 = 12.1\n"
    "When h = 0.00001: slope  (8.00012 - 8) / 0.00001 = 12.0006\n"
)

# Define "x" and step size "h"
x, h = symbols('x h')
# Function
f_sym = x**3
# substitute into rise-over-run formula
slope_f = (f_sym.subs(x, x + h) - f_sym) / h
# x = 2
slope_2 = slope_f.subs(x, 2)
# infinitely approach step size _h_ to 0
result = limit(slope_2, h, 0)

print(f"The limit as h approaches 0 is: {result}")

f_prime = diff(f_sym, x)
display(Latex(f"The derivative of $f(x) = x^3$ using the power rule is: ${latex(f_prime)}$. The derivative at x=2 is: ${latex(f_prime.subs(x, 2))}$"))
display(Latex(f"The rate of change (at any point): $ {latex(limit(slope_f, h, 0))} $"))

```

```python
# Partial derivatives (multivariable functions)

print("Partial derivatives for multivariable functions")
display(Latex(f"$f(x, y) = 2x^3 + 3y^3$"))

x,y = symbols('x y')
f = 2*x**3 + 3*y**3
# Calculate the partial derivatives for x and y using SymPy:
dx_f = diff(f, x)
dy_f = diff(f, y)

display(Latex(f"$ \\frac{{\\partial}}{{\\partial x}} f(x, y) = {latex(dx_f)} $"))
display(Latex(f"$ \\frac{{\\partial}}{{\\partial y}} f(x, y) = {latex(dy_f)} $"))
plot3d(f)
```

```python
# The Chain Rule (finding the derivative of composite functions)

display(Latex(f"$y = x^2 + 1$"))
display(Latex(f"$z = y^3 - 2$"))
print("Above two functions are linked (the 'y' is the output variable in the first function but is the input variable in the second)")
display(Latex(f"$z = (x^2 + 1)^3 - 2$"))

display(Latex(f"Finding the derivative of $z$ with respect to $x$"))

x, y = symbols('x y')
z = (x**2 + 1)**3 - 2
dz_dx = diff(z, x)
display(Latex(f"$\\frac{{dz}}{{dx}} = {latex(dz_dx)}$"))

print("Derivatives of the y and z functions separately, and then multiply them together")
display(Latex(f"$\\frac{{dz}}{{dx}} = \\frac{{dy}}{{dx}} \\times \\frac{{dz}}{{dy}}$"))
display(Latex(f"$\\frac{{dy}}{{dx}} (x^2 + 1) = {latex(diff(x**2 + 1, x))}$"))
display(Latex(f"$\\frac{{dz}}{{dy}} (y^3 - 2) = {latex(diff(y**3 - 2, y))}$"))
display(Latex(
    f"$\\frac{{dz}}{{dx}} = ({latex(diff(x**2 + 1, x))})({latex(diff(y**3 - 2, y))}) "
    f"= {latex(diff(x**2 + 1, x) * diff(y**3 - 2, y))} = {latex(dz_dx)}$"))


display(Latex("Sympy - Calculating the derivative $\\frac{{dz}}{{dx}}$ with and without the chain rule, but still getting the same answer:"))
_y = x**2 + 1
_z = y**3 - 2
dy_dx = diff(_y)
dz_dy = diff(_z)
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(_z.subs(y, _y))
display(Latex(
    f"Using the chain rule: $\\frac{{dz}}{{dx}} = {latex(dz_dx_chain)}$ \n"
    f"Without using the chain rule: $\\frac{{dz}}{{dx}} = {latex(dz_dx_no_chain)}$"
))
```
